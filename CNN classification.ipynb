```python
#First I convert all collected images into JPG format.
input_dir = r'C:\Users\dell i7\Desktop\Photo\cow'
output_dir = r'C:\Users\dell i7\Desktop\Photo\cowJPG'

os.makedirs(output_dir, exist_ok=True)

input_files = os.listdir(input_dir)

for input_file in input_files:
    input_path = os.path.join(input_dir, input_file)

    if os.path.isfile(input_path):
        try:
            with Image.open(input_path) as img:
                img = img.convert("RGB")
                output_file = os.path.splitext(input_file)[0] + ".jpg"
                output_path = os.path.join(output_dir, output_file)
                img.save(output_path, "JPEG")
                print(f"Converted {input_file} to {output_file}")
        except Exception as e:
            print(f"An error occurred while processing {input_file}: {str(e)}")
```

```python
# Create a dataframe for each folder category of the object data set
# convert the file paths and labels into pandas series. organize the data into row and column also make easy for exploration.

from pathlib import Path
import pandas as pd

train_path = r'C:\Users\dell i7\Desktop\Image dataset\Image classification\train'
train_dir = Path(train_path)
filepaths = list(train_dir.glob(r'**/*.JPG'))
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))
filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')
train_df = pd.concat([filepaths, labels], axis=1)

val_path = r'C:\Users\dell i7\Desktop\Image dataset\Image classification\validation'
val_dir = Path(val_path)
filepaths = list(val_dir.glob(r'**/*.JPG'))
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))
filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')
val_df = pd.concat([filepaths, labels], axis=1)

test_path = r'C:\Users\dell i7\Desktop\Image dataset\Image classification\test'
test_dir = Path(test_path)
filepaths = list(test_dir.glob(r'**/*.JPG'))
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))
filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')
test_df = pd.concat([filepaths, labels], axis=1)
```

```python
# Creating our train and validation generators in the absence of any data augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Rescale all images by 1/255. 
# Neural Networks work better with smaller input values.
train_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_dataframe(train_df,
                                                   x_col='Filepath',
                                                   y_col='Label',
                                                   target_size = (150, 150),
                                                   batch_size = 20,
                                                   shuffle = True
                                                   )

validation_generator = test_datagen.flow_from_dataframe(val_df,
                                                       x_col='Filepath',
                                                       y_col='Label',
                                                       target_size = (150, 150),
                                                       batch_size = 20
                                                       )

test_generator = test_datagen.flow_from_dataframe(test_df,
                                                  x_col='Filepath',
                                                  y_col='Label',
                                                  target_size = (150, 150),
                                                  batch_size = 20,
                                                  shuffle = False
                                                    )
```

# Object Recognition Model

```python
# instantiate a baseline CNN for Object Classification
from tensorflow.keras import models, layers

# adding a base CNN network
# convolutional layers for learning features and max-pooling are reduce the dimensions
# Rectified linier unit are activation function for choose max value for extract good features.
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (150,150,3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation = 'relu'))
model.add(layers.MaxPooling2D((2, 2)))

## adding a classifier. there are 10 classes of image so I use softmax because it is use of multi class classification.
model.add(layers.Flatten())
model.add(layers.Dense(10, activation = 'relu'))
model.add(layers.Dense(10, activation = 'softmax'))

model.summary()
# channel represent features extraction from the input image. extract 32 different features or patterns from the input data.
# param is total number of numeric values that the model will learn during training. sum of weight and bia
```

```python
# compile the model for training
# the learning rate halps the model converge faster. smaller learning rates to slover but stable convergence.
from tensorflow.keras import optimizers

model.compile(loss = 'categorical_crossentropy'
             ,optimizer = optimizers.RMSprop(learning_rate = 1e-4)
             ,metrics = ['acc'])
```

```python
history = model.fit(train_generator
         ,steps_per_epoch = 100
         ,epochs = 100
         ,validation_data = validation_generator
         ,validation_steps = 50)
```

```python
# Plot a learning curve.
# The learning curve will help us monitor if our baseline model is overfitting or underfitting.

import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc)+1)

plt.plot(epochs, acc, 'bo', label = 'Training Acc')
plt.plot(epochs, val_acc, 'r', label = 'Validation Acc')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label = 'Training Loss')
plt.plot(epochs, val_loss, 'r', label = 'Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.show()
```

# Fine-Tuning for Object Categories

```python
# get the Inception V3 base model
# the sequential model used for feedforward nural network. when the dataflow is one direction.
# inception_v3 pre-trained model developed by google. used for excels at capturing complex and hierarchial features in images.
from tensorflow.keras.applications import InceptionV3

base_model = InceptionV3(include_top = False, input_shape = (150, 150, 3))
base_model.trainable = False

base_model.summary()
```

```python
from tensorflow.keras import layers, models

# define the hidden layers
# 'mixed10' is typically one of the last layers in the InceptionV3 architecture, and it contains features that are suitable for high-level representation.
x = base_model.get_layer('mixed10').output # mixed10 is the last layer of the base_model
x = layers.GlobalAveragePooling2D()(x) # global average is preferred over flatten becuase the lesser variables makes the network more efficient

# define the output layer
outputs = layers.Dense(10, activation = 'softmax')(x)

# create the model
model = models.Model(base_model.input, outputs)
model.summary()
```

```python
# compile the model for training
# Adam is a popular optimization algorithm that adapts the learning rates for each parameter during training, making it suitable for a wide range of deep learning tasks.
# The learning rate (0.01 in this case) controls the step size of weight updates. 
from tensorflow.keras import optimizers

model.compile(loss = 'categorical_crossentropy'
             ,optimizer = optimizers.Adam(learning_rate = 0.01)
             ,metrics = ['acc'])
```

```python
# training with flattened CNN
history = model.fit(train_generator
         ,steps_per_epoch = len(train_generator)
         ,epochs = 10
         ,validation_data = validation_generator
         ,validation_steps = int(0.25*len(validation_generator)))
```

```python
model.trainable = True
# Un-freeze last 4 layers
for layer in model.layers[:-4]:
    layer.trainable = False
```

```python
import numpy as np

# use the model to predict all test images
pred = model.predict(test_generator)
pred = np.argmax(pred, axis = 1)

# Map the label
labels = (test_generator.class_indices)
labels = dict((v,k) for k,v in labels.items())
pred = [labels[k] for k in pred]
```

```python
import matplotlib.pyplot as plt

# Display 40 random pictures from the dataset with their labels
random_index = np.random.randint(0, len(test_df) - 1, 40)
fig, axes = plt.subplots(nrows=8, ncols=5, figsize=(25, 15),
                        subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(axes.flat):
    ax.imshow(plt.imread(test_df.Filepath.iloc[random_index[i]]))
    if test_df.Label.iloc[random_index[i]] == pred[random_index[i]]:
        color = "green"
    else:
        color = "red"
    ax.set_title(f"True: {test_df.Label.iloc[random_index[i]]}\nPredicted: {pred[random_index[i]]}", color=color)
    
correct_predictions = sum(1 for i in random_index if test_df.Label.iloc[i] == pred[i])
total_samples = len(random_index)
accuracy = correct_predictions / total_samples * 100

plt.suptitle(f"Model Accuracy: {accuracy:.2f}%", fontsize=16, y=0.95)    
plt.tight_layout()
plt.show()
```
